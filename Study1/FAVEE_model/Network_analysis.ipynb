```r
library(visNetwork)
library(tidyverse)
#library(pandoc)
#library(webshot)
```

# Import data

```r
FAVEE = c('Formality','Activeness','Valence','Exchange','Equality')
FAVEE_single = c('Formality and Regulation','Activeness','Valence Evaluation',
              'Money Exchange','Equality')
```

```r
USA_scaled = read.csv('output_data/cleaning_results/USA_dim_rel_scaled_30d.csv',
                     row.names=1,check.names=FALSE)
USA_scaled.corr = abs(cor(USA_scaled,method='pearson'))
```

# Prepare edge/node

```r
generate_edge = function(raw_scaled.corr){
    edges = data.frame(from=character(),to=character(),value=numeric())
    count = 0
    for (dim1 in rownames(raw_scaled.corr)){
        for (dim2 in colnames(raw_scaled.corr)){
            if (dim1 != dim2){
                edges[count,] = c(dim1,dim2,raw_scaled.corr[dim1,dim2])
                count = count + 1
            } 
        }
    }
    return(edges)
}

generate_node = function(raw_scaled.corr){
    nodes = data.frame(id=character(30),label=character(30))#,
                       #value=numeric(30))
    nodes$'id' = rownames(raw_scaled.corr)
    nodes$'label' = rownames(raw_scaled.corr)
    #nodes$'value' = 0.1
    #nodes[nodes$'label'%in%FAVEE_single,'value']=0.15
    #for (dim in rownames(raw_scaled.corr)){
        #sum_cor = sum(raw_scaled.corr[dim,])-1 # delete the diagonal
        #avg_cor = sum_cor/32
        #nodes[nodes$'id'==dim,'value'] = sum_cor
    #}
    return(nodes)
}
```

```r
USA_edges = generate_edge(USA_scaled.corr)
head(USA_edges)
USA_nodes = generate_node(USA_scaled.corr)
head(USA_nodes)
```

# Plot results

```r
edges_filter = USA_edges %>% 
    filter(from%in%FAVEE_single) %>% filter(value>0.4)

#head(edges_filter)

nodes_filter = USA_nodes %>% 
    mutate(font.size = case_when(id %in% FAVEE_single~30,
           TRUE~20)) %>%
    mutate(group = case_when(id=='Formality and Regulation'~'Formality',
                            id=='Activeness'~'Activeness',
                            id=='Valence Evaluation'~'Valence',
                            id=='Money Exchange'~'Exchange',
                            id=='Equality'~'Equality'),
          color=case_when(id=='Formality and Regulation'~'#009E73',
                          id=='Activeness'~'#CC79A7',
                            id=='Valence Evaluation'~'#D55E00',
                            id=='Money Exchange'~'#F0E442',
                            id=='Equality'~'#56B4E9',
                         TRUE~'#D0CECE'))
head(nodes_filter)

vn = visNetwork(nodes_filter, edges_filter,
           background='white',
           height = "1000px", width = "60%")%>%
    visNodes(size=10)%>%
    visLayout(randomSeed = 2023) 
vn
```
