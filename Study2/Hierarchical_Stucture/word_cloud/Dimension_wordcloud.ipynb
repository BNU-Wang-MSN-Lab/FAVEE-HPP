```python
import pandas as pd

from wordcloud import WordCloud, STOPWORDS
from PIL import Image

#import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
def data_org_dict(df):
    result = {col: [] for col in df.columns}
    
    for index, row in df.iterrows():
        max_col = row.abs().idxmax()
        result[max_col].append(index)
    
    return result
```

```python
def wc(level,dim,w,h):
    # put sailent feature in each component
    dim_df = pd.read_csv(f'level_results/15regions_2385r/Forbes_level{level}.csv', index_col=0)
    dim_df.index = dim_df.index.str.replace('.',' ', regex=False)
    dim_dict = data_org_dict(dim_df)
    dimensions = dim_dict[dim]
    print(len(dimensions))
    score_dict = dim_df.loc[dimensions,dim].abs().to_dict()
    # plot wordcloud
    wordcloud = WordCloud(width = w, height = h,
                font_path=r"C:\Windows\Fonts\Arial.ttf",
                background_color ='#F2F2F2', 
                max_font_size=60,
                mode='RGBA',color_func=lambda *args, **kwargs: "black")
    wordcloud.generate_from_frequencies(score_dict)
    plt.figure(dpi=150)
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.show()
    wordcloud.to_file(dim+str(level)+".png")
```
