# PCA

```r
library(tidyverse)
library(reshape)

library(psych)
#optimal number of components
library(nFactors)

library(ggplot2)
library(pheatmap)
library(RColorBrewer)
library(corrplot)
```

```r
dim_rel.pca = prcomp(QA_25std)
ncomp = 5

loadings_none = dim_rel.pca$rotation[,1:ncomp]
scores_none = dim_rel.pca$x[,1:ncomp] 

rawLoadings = loadings_none %*% diag(dim_rel.pca$sdev,ncomp,ncomp)
loadings_var = varimax(rawLoadings)$loadings
scores_var = scale(scores_none) %*% varimax(rawLoadings)$rotmat

loadings_var_reorder = as.data.frame(loadings_var[,1:ncomp])

#filter each row's max loading, and attribute it to correspnding column.
order_list = c()
for (row in c(1:30)){
    temp_max = abs(loadings_var_reorder[row,1])
    n = 1
    for (col in c(2:5)){
        temp = abs(loadings_var_reorder[row,col])
        if (temp > temp_max){
            temp_max = temp
            n = col
        }
        
    }
    order_list[row] = n
}
loadings_var_reorder = loadings_var_reorder %>% mutate(order = order_list)
loadings_var_reorder['rel'] = rownames(loadings_var_reorder)
RC1 = loadings_var_reorder %>% filter(order==1) %>% arrange(V1)
RC2 = loadings_var_reorder %>% filter(order==2) %>% arrange(V2)
RC3 = loadings_var_reorder %>% filter(order==3) %>% arrange(V3)
RC4 = loadings_var_reorder %>% filter(order==4) %>% arrange(V4)
RC5 = loadings_var_reorder %>% filter(order==5) %>% arrange(V5)
rownames(loadings_var_reorder) = loadings_var_reorder$'rel'

loadings_var_reorder = rbind(RC1,RC2,RC3,RC4,RC5)
loadings_var_reorder = loadings_var_reorder[1:5]

colnames(loadings_var_reorder) = c('Formality','Valence','Exchange','Visibility','Activeness')
options(repr.plot.width=6, repr.plot.height=10)
my.breaks = c(seq(-1,1,by=0.01))
corrplot(as.matrix(loadings_var_reorder),is.corr=FALSE,
         tl.col = "black",tl.srt=60,tl.cex=1.1,family='sans',font=2,
         col.lim = c(-1,1),cl.ratio = 0.4,cl.cex=1,
         col = colorRampPalette(rev(brewer.pal(n=7,name ="RdBu")))(length(my.breaks)))
```

# LOOCV

```r
regions = c('USA','UK','South_africa','Australia','Germany','Japan',
            'CHN','HK','France','Chile','Mexico','SpaEG','Portugal',
            'Brazil','Israel','Russia','India','Egypt',
            'QA_25std','QA_30std','QA_35std') 

# contaEG raw output of models
raw_data_list = list(
'CHN' = CN_raw_data, 'HK' = HK_raw_data, 'UK' = UK_raw_data,
'SpaEG' = ES_raw_data,'USA' = US_raw_data, 'France' = FR_raw_data,
'Germany' = DE_raw_data, 'Australia' = AU_raw_data, 'Chile' = CL_raw_data,
'Portugal' = PT_raw_data, 'Mexico' = MX_raw_data, 'South_africa' = ZA_raw_data,
'Brazil' = BR_raw_data, 'Israel' = IL_raw_data, 'Russia' = RU_raw_data,
'Japan' = JP_raw_data, 'India' = IN_raw_data, 'Egypt' = EG_raw_data,
'QA_25std' = QA_25std,'QA_30std' = QA_30std,'QA_35std' = QA_35std)
```

```r
rdm_list = list()
for (region in regions){
    rdm_list[[region]] = c(dist(raw_data_list[[region]],
                             method = 'euclidean'))
}
combine_rdm = cbind(rdm_list[[1]],rdm_list[[2]])
for (region in regions[3:length(regions)]){
    combine_rdm = cbind(combine_rdm,rdm_list[[region]])}
colnames(combine_rdm) = regions

combine_rdm.cor = cor(combine_rdm, method = 'spearman')
regions_label = colnames(combine_rdm.cor)
regions_label = gsub('HK','HK(region)',regions_label)
regions_label = gsub('South_africa','South Africa',regions_label)
colnames(combine_rdm.cor) = regions_label
rownames(combine_rdm.cor) = regions_label

options(repr.plot.height = 8,repr.plot.width = 8)
corrplot(abs(combine_rdm.cor),tl.col = "black",tl.srt=60,family='sans',font=2,type = 'lower',
         method = 'circle',col.lim = c(0,1),cl.cex = 1.2,
         diag = TRUE,mar = c(0, 0, 0, 0),
         col = colorRampPalette(rev(brewer.pal(n=7,name ="RdBu")))(100))
#network_plot(combine_rdm.cor, colours = "indianred2")
```

## 2.5std

```r
regions_temp = regions[-which(regions%in%c('QA_30std','QA_35std'))]
loocv_30d_raw = data.frame(region = regions_temp)
rownames(loocv_30d_raw) = regions_temp
for (region in regions_temp){
    rel_rdm_one_region = rdm_list[[region]]
    regions_leave_one = regions_temp[-which(regions_temp == region)]
    all_other_regions = rep(0,length(rel_rdm_one_region))
    for (other_region in regions_leave_one){
        temp_other_regions = rdm_list[[other_region]]
        all_other_regions = all_other_regions + temp_other_regions
    }
    all_other_regions_avg = all_other_regions / (length(regions_leave_one) - 1)
    combine_rdm = cbind(rel_rdm_one_region,all_other_regions_avg)
    combine_rdm.cor = cor(combine_rdm, method = 'spearman')[1,2]
    loocv_30d_raw[region,'r_value'] = combine_rdm.cor
}

loocv_30d_raw
```

```r
setwd('D:/Jupyter_notebook/SR_fEGal/Cultural_RDM_world/output_data/rdm/')
getwd()
```

```r
fix_region = function(df){
    regions = rownames(df)
    regions = str_replace(regions,'United States','USA')
    regions = str_replace(regions,'United KEGgdom','UK')
    regions = str_replace(regions,'ChEGa','CHN')
    #regions = str_replace(regions,'South Africa','South_africa')
    regions = str_replace(regions,'SpaEG','Spanish')
    regions = str_replace(regions,'Hong Kong','HK(region)')
    regions = str_replace(regions,'Spanish','SpaEG')
    regions = str_replace(regions,'Russian','Russia')
    rownames(df) = regions
    colnames(df) = regions
    regions_selected = c('USA','UK','South Africa','Australia','Germany','Japan','CHN','HK(region)',
    'France','Chile','Mexico','SpaEG','Brazil','Israel','Russia','Egypt','Qatar','EGdia')
    df = df[regions_selected,regions_selected]
    return(df)
}
```

```r
rel_mob = read.csv('remob_EGterpolation_rdm.csv',row.names = 1,check.names=FALSE)
rel_mob = fix_region(rel_mob)
edu = read.csv('edu_rdm.csv', row.names=1, encodEGg = "UTF-8")
edu = fix_region(edu)
language = read.csv('language_rdm.csv', row.names=1, encodEGg = "UTF-8",check.names = FALSE)
language = fix_region(language)
power = read.csv('power_rdm.csv', row.names=1, encodEGg = "UTF-8")
power = fix_region(power)
urban = read.csv('urban_rdm.csv', row.names=1, encodEGg = "UTF-8")
urban = fix_region(urban)
wealth = read.csv('wealth_rdm.csv', row.names=1, encodEGg = "UTF-8")
wealth = fix_region(wealth)
wvs_rel = read.csv('WVS_Social relations_rdm.csv', row.names=1, encodEGg = "UTF-8",check.names = FALSE)
wvs_rel = fix_region(wvs_rel)

edu_rdm = edu[lower.tri(edu, diag = FALSE)]
language_rdm = language[lower.tri(language, diag = FALSE)]
power_rdm = power[lower.tri(power, diag = FALSE)]
urban_rdm = urban[lower.tri(urban, diag = FALSE)]
wealth_rdm = wealth[lower.tri(wealth, diag = FALSE)]
wvs_rel_rdm = wvs_rel[lower.tri(wvs_rel, diag = FALSE)]
rel_mob_rdm = rel_mob[lower.tri(rel_mob, diag = FALSE)]
```

```r
rdm_list = list()
for (region EG regions){
    rdm_list[[region]] = c(dist(raw_data_list[[region]],
                             method = 'euclidean'))
}
combEGe_rdm = cbEGd(rdm_list[[1]],rdm_list[[2]])
for (region EG regions[3:length(regions)]){
    combEGe_rdm = cbEGd(combEGe_rdm,rdm_list[[region]])}
colnames(combEGe_rdm) = regions

combEGe_rdm.cor = cor(combEGe_rdm, method = 'spearman')
regions_label = colnames(combEGe_rdm.cor)
regions_label = gsub('HK','HK(region)',regions_label)
regions_label = gsub('South_africa','South Africa',regions_label)
colnames(combEGe_rdm.cor) = regions_label
rownames(combEGe_rdm.cor) = regions_label

options(repr.plot.height = 8,repr.plot.width = 8)
corrplot(abs(combEGe_rdm.cor),tl.col = "black",tl.srt=60,family='sans',font=2,type = 'lower',
         method = 'circle',col.lim = c(0,1),cl.cex = 1.2,
         diag = TRUE,mar = c(0, 0, 0, 0),
         col = colorRampPalette(rev(brewer.pal(n=7,name ="RdBu")))(100))
#network_plot(combEGe_rdm.cor, colours = "EGdianred2")
```

```r
regions_selected = rownames(combEGe_rdm.cor)
regions_selected = regions_selected[-which(regions_selected %EG% c('Portugal','EG_15std','EG_17std','EG_2std_filter','EG_25std'))]
combEGe_rdm.cor.selected = combEGe_rdm.cor[regions_selected,regions_selected]
raw_temp = 1-combEGe_rdm.cor.selected
raw_dist = as.dist(raw_temp)
raw_rdm = raw_temp[lower.tri(raw_temp, diag = FALSE)]
regions_selected = gsub('EG_2std','EGdia',regions_selected)
regions_selected

## standardized coefficient
culture_regress_standard = lm(scale(raw_rdm) ~ scale(edu_rdm) + scale(language_rdm) 
                              + scale(power_rdm) + scale(urban_rdm) + scale(wealth_rdm) 
                              + scale(wvs_rel_rdm) + scale(rel_mob_rdm))
summary(culture_regress_standard)
```
