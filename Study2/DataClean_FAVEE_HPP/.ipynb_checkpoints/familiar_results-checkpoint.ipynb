```python
import pandas as pd
import glob
import re

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def cluster_results(result_df, method_number='cl_k5'):
    for n in range(len(result_df[method_number].unique())):
        each_cluster = result_df.index[result_df[method_number]==n]
        print("\nCluster "+str(n+1)+" Relationship:{}".format(len(each_cluster)))
        print(*result_df.index[result_df[method_number]==n], sep = ", ")
```

# familiar results

```python
relationships = pd.read_csv('input_data/159relationships_fix_label.csv',index_col=0)
relationships = relationships.iloc[:,0].to_list()

relationships_258 = pd.read_csv('input_data/258relationships_fix_label.csv',index_col=0)
relationships_258 = relationships_258.iloc[:,0].to_list()

files_path = glob.glob('output_data/check_quality/familiarity/*.csv')
# abstract regions we selected
regions = [re.search('(?<=familiarity\\\\)(.*).csv',path).group(1) for path in files_path]
```

```python
region_relationship_df = pd.DataFrame(index=regions, columns=relationships)
for reg_path in files_path:
    region = reg_path.split('\\')[1].split('.csv')[0]
    region_df = pd.read_csv(reg_path,index_col = 0)
    region_df = region_df.apply(pd.to_numeric)
    
    # different regions contain different columns
    results = re.search('CHN',reg_path)
    if results:
        region_df = region_df.iloc[:,:-1]
        region_df.columns = relationships_258
        region_df = region_df.loc[:,relationships]
    
    elif len(region_df.columns) == 160: # some regions contain subjects'ID
        region_df = region_df.iloc[:,:-1]
        region_df = region_df.apply(pd.to_numeric)
        
    region_df.columns = relationships
    # caculate mean results
    region_mean_df = region_df.mean(axis=0)
    region_relationship_df.loc[region,] = region_mean_df

# all regions' results
world_df = pd.DataFrame(region_relationship_df.mean(),columns=['familiar'])
```

# 6 clusters results in Study2

## plot familiar results

```python
umap_exp_6k = pd.read_csv('../../Study2/output_data/group/category/cluster_results/exp/umap_exp_results_159r_all.csv',
                                       index_col=0)
fix_rels = pd.read_excel('../../Study2/input_data/fix_labels_wordcloud.xlsx')
fix_rels.columns = ['raw','new']
fix_rels_zipped = zip(fix_rels['raw'], fix_rels['new'])
fix_rels_map = {key:value for key, value in fix_rels_zipped}
umap_exp_6k.index = umap_exp_6k.index.map(fix_rels_map)


fix_familiarity_rels = pd.read_excel('../../PCA/output_data/fix_label_familiarity.xlsx',
                                   header = None)
fix_familiarity_rels.columns = ['raw','new']
# fix_familiarity_rels_zipped = zip(fix_familiarity_rels['raw'], 
#                                   fix_familiarity_rels['new'])
fix_familiarity_rels_zipped = zip(fix_familiarity_rels['new'], 
                                  fix_familiarity_rels['raw'])
fix_familiarity_rels_map = {key:value for key, value in fix_familiarity_rels_zipped}
umap_exp_6k.index = umap_exp_6k.index.map(fix_familiarity_rels_map)
```

```python
cluster_results(umap_exp_6k, method_number='cl_k6')
```
