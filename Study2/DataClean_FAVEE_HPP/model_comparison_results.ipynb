```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
model_rank_regions = pd.read_csv('output_data/compare_model/model_rank_regions.csv', index_col = 0)
model_rank_regions = model_rank_regions.drop('Region', axis=1)
model_rank_regions.columns
```

```python
model_rank = model_rank_regions.groupby('Model').mean()
model_rank_sorted = model_rank.sort_values(by='BIC', ascending = False)
model_rank_regions['Model'] = pd.Categorical(model_rank_regions['Model'], categories=model_rank_sorted.index, ordered=True)

plt.figure(figsize=(25,18))
sns.set(context='notebook', style='white', font_scale=3)

ax = sns.barplot(x='Model', y='BIC', data=model_rank_regions, order=model_rank_sorted.index, errorbar = ('se'),
                 facecolor="#F8D6BA", edgecolor=".2", linewidth=7.5, errwidth=7.5)
sns.scatterplot(x='Model', y='BIC', data=model_rank_regions, color='gray', s=300, zorder=10, ax=ax, alpha=0.5)

ax.set_ylim(350, 440)

ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right", fontsize=10)
ax.set_yticklabels(ax.get_yticklabels(), rotation=270, ha="right", fontsize=50)
ax.invert_yaxis()

for spine in ax.spines.values():
    spine.set_linewidth(7.5)

# ax.set_xlim(ax.get_xlim()[0]-0.5, ax.get_xlim()[1]+0.5)  
plt.show()
```
