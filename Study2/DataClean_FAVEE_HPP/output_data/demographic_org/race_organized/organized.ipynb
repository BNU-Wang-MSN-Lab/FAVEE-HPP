# combine enthnicity

```python
others10 = pd.read_excel('Ethnicity.xlsx',sheet_name='others',index_col=0)
CHN_HK = pd.read_excel('Ethnicity.xlsx',sheet_name='CHN_HK',index_col=0)
Japan = pd.read_excel('Ethnicity.xlsx',sheet_name='Japan',index_col=0)
Russia = pd.read_excel('Ethnicity.xlsx',sheet_name='Russia',index_col=0)
Germany = pd.read_excel('Ethnicity.xlsx',sheet_name='Germany',index_col=0)
Brazil = pd.read_excel('Ethnicity.xlsx',sheet_name='Brazil',index_col=0)
Portugal = pd.read_excel('Ethnicity.xlsx',sheet_name='Portugal',index_col=0)
Mexico = pd.read_excel('Ethnicity.xlsx',sheet_name='Mexico',index_col=0)
Chile = pd.read_excel('Ethnicity.xlsx',sheet_name='Chile',index_col=0)
regions = others10.join([CHN_HK,Japan,Russia,Germany,Brazil,Portugal,Mexico,Chile],how='outer')
```

```python
regions.loc['American Indian or Alaska Native',:] = np.nan
regions.loc['Native Hawaiian or Pacific Islander',:] = np.nan
print(regions.shape)
regions=regions.loc[['White', 'Black', 'South Asian Indian, Pakistani, etc.',
       'East Asian Chinese, Japanese, etc.', 'Arabic, Central Asian',
        'Asian','Han', 'Yamato Ethnic Group', 'Ryukyu Peoples',
       'Ainu Ethnicity', 'Russian', 'Tatar', 'Ukrainian', 'Belorussian ',
       'Jewish', 'German', 'Dane', 'Sorbian', 'Brown or mulatto', 'Indigenous',
       'Portuguese', 'Latino', 'Mestizo', 'Mulatto',
       'American Indian or Alaska Native',
       'Native Hawaiian or Pacific Islander','Other (write in)'],:]
regions.to_csv('organized_ethnicity.csv')
regions
```
