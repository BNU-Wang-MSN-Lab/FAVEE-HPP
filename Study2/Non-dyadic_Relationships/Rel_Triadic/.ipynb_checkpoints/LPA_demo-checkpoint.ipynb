```r
library(dplyr)
library(mclust)
library(reshape2)
library(data.table)
library(tidyr)
library(tibble)
library(ggplot2)
```

```r
relationships_scaled<-read.csv("output_data/cleaning_results/triadic_dim_rel_scaled_33d_all.csv")
relationships_clean <- relationships_scaled[c(1:35), c(2:34)]
relationships_clustering <- relationships_clean %>%
  na.omit() %>%
  mutate_all(list(scale))
```

```r
BIC <- mclustBIC(relationships_clustering)
plot(BIC)
summary(BIC)
```

```r
ICL <- mclustICL(relationships_clustering)
plot(ICL)
summary(ICL)
```

```r
mclustBootstrapLRT(relationships_clustering, modelName = "VEE")
```

```r
mod1 <- Mclust(relationships_clustering,  G = 3)
summary(mod1)
```

```r
means <- data.frame(mod1$parameters$mean, stringsAsFactors = FALSE) %>%
  rownames_to_column() %>%
  rename(EvaluativeFeatures = rowname) %>%
  melt(id.vars = "EvaluativeFeatures", variable.name = "Profile", value.name = "Mean") %>%
  mutate(Mean = round(Mean, 2),
         Mean = ifelse(Mean > 1, 1, Mean))
```

```r
means %>%
  ggplot(aes(EvaluativeFeatures, Mean, group = Profile, color = Profile)) +
  geom_point(size = 2.25) +
  geom_line(size = 1.25) +
  scale_x_discrete(limits = c("Valence Evaluation", "Equality", "Activity Intensity",
                              "Socioemotional", "Formality and Regulation", "Intimacy",
                              "Visibility", "Communal Sharing", "Expected Reciprocity", "Strategic", "Concreteness", "Uniqueness",
                              "Love Expression", "Information Exchange", "Money Exchange", "Goods Exchange", "Service Exchange", "Attachment", "Mating",
                              "Conflict", "Negotiation", "Coercion", "Importance for society", "Importance for individuals", "Affiliation Coalition",
                              "Occupational", "Activeness", "Endurance", "Synchronicity", "Spatial Distance", "Morality", "Trust", "Generation Gap")) +
  labs(x = NULL, y = "Standardized mean EF") +
  theme_bw(base_size = 14) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1), legend.position = "top")
```
