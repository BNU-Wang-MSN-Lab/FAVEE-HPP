```r
library(dplyr)
library(mclust)
library(reshape2)
library(data.table)
library(tidyr)
library(tibble)
library(ggplot2)
```

```r
relationships_scaled<-read.csv("output_data/cleaning_results/CHN_dim_rel_scaled_NLP_159r_33d.csv")
relationships_clean <- relationships_scaled[c(1:160), c(2:34)]
relationships_clustering <- relationships_clean %>%
  na.omit() %>%
  mutate_all(list(scale))
```

```r
BIC <- mclustBIC(relationships_clustering)
plot(BIC)
summary(BIC)
```

```r
ICL <- mclustICL(relationships_clustering)
plot(ICL)
summary(ICL)
```

```r
mclustBootstrapLRT(relationships_clustering, modelName = "VEE")
```

```r
mod1 <- Mclust(relationships_clustering,  G = 7)
summary(mod1)
```

```r
print(mod1)
```

```r
# Calculate Entropy for the mod1 model
calculate_entropy <- function(prob_matrix) {
  entropy <- 0
  for (i in 1:nrow(prob_matrix)) {
    for (k in 1:ncol(prob_matrix)) {
      p_ik <- prob_matrix[i, k]
      if (p_ik > 0) {  # To avoid log2(0)
        entropy <- entropy - p_ik * log2(p_ik)
      }
    }
  }
  return(entropy)
}

# Extract the probability matrix from the mod1 model
prob_matrix <- mod1$z

# Calculate the Entropy
entropy_mod1 <- calculate_entropy(prob_matrix)

# Print the Entropy
print(paste("Entropy of mod1 model:", entropy_mod1))
```

```r
means <- data.frame(mod1$parameters$mean, stringsAsFactors = FALSE) %>%
  rownames_to_column() %>%
  rename(EvaluativeFeatures = rowname) %>%
  melt(id.vars = "EvaluativeFeatures", variable.name = "Profile", value.name = "Mean") %>%
  mutate(Mean = round(Mean, 2),
         Mean = ifelse(Mean > 1, 1, Mean))
```

```r
means %>%
  ggplot(aes(EvaluativeFeatures, Mean, group = Profile, color = Profile)) +
  geom_point(size = 2.25) +
  geom_line(size = 1.25) +
  scale_x_discrete(limits = c("Formality and Regulation","Occupational","Strategic","Visibility","Importance for society",
                              "Socioemotional", "Attachment","Mating","Love Expression", 
                              "Activeness","Importance for individuals","Synchronicity", "Activity Intensity", 
                              "Uniqueness","Information Exchange","Endurance", "Intimacy","Spatial Distance",
                              "Valence Evaluation","Morality", "Trust", "Affiliation Coalition", "Conflict","Coercion", 
                              "Money Exchange", "Goods Exchange", "Concreteness", "Service Exchange", "Negotiation", "Communal Sharing",  
                               "Equality", "Expected Reciprocity", "Generation Gap")) +
  labs(x = NULL, y = "Standardized mean EF") +
  theme_bw(base_size = 14) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1), legend.position = "top")
```
