# correlation

## 30d

### relationship score

```r
USA_scores = read.csv('output_data/pca_results/USA/USA_scores_var_30d.csv',
                       row.names=1)
colnames(USA_scores) = c('Formality','Activeness','Valence','Exchange','Equality')
USA_scores['Equality'] = -USA_scores['Equality']
USA_scores = USA_scores[,FAVEE]

CHN_scores = read.csv('output_data/pca_results/CHN/CHN_NLP_258rel_30d_scores_var.csv',
                       row.names=1)
colnames(CHN_scores) = c('Activeness','Exchange','Formality','Valence','Equality')
CHN_scores['Activeness'] = -CHN_scores['Activeness']
relationships_159 = read.csv('../DataCleanPCA/input_data/RelDim_labels/159relationships_fix_label.csv',row.names=1)
relationships_159 = relationships_159$'X0'
CHN_scores = CHN_scores[relationships_159,FAVEE]

Trans_scores = read.csv('output_data/pca_results/CHN/CHN_Trans_159rel_30d_scores_var.csv',
                       row.names=1)
colnames(Trans_scores) = c('Valence','Exchange','Formality','Activeness','Equality')
Trans_scores['Formality'] = -Trans_scores['Formality']
Trans_scores['Valence'] = -Trans_scores['Valence']
Trans_scores['Equality'] = -Trans_scores['Equality']
Trans_scores = Trans_scores[,FAVEE]
```

```r
USA_scores_corr = USA_scores
colnames(USA_scores_corr) = c('Formality(USA)','Activeness(USA)','Valence(USA)','Exchange(USA)','Equality(USA)')

CHN_scores_corr = CHN_scores
colnames(CHN_scores_corr) = c('Formality(CHN)','Activeness(CHN)','Valence(CHN)','Exchange(CHN)','Equality(CHN)')

Trans_scores_corr = Trans_scores
colnames(Trans_scores_corr) = c('Formality(Trans)','Activeness(Trans)','Valence(Trans)',
                                  'Exchange(Trans)','Equality(Trans)')

options(repr.plot.width=10)
Corr_all = cbind(USA_scores_corr,CHN_scores_corr,Trans_scores_corr)
Corr = cor(Corr_all, method = "spearman")
corrplot(abs(Corr), tl.col="black", 
         tl.srt=0.001, cl.offset = 1, tl.cex=1.2, col.lim = c(0,1), 
         col= colorRampPalette(c("midnightblue","white","darkred"))(100), 
         outline=TRUE, addgrid.col="white", type="lower", diag=T)

# no labels
png("../output_graph/USA_CHN/similarity_30d.png", bg="transparent",
    family = 'sans',units='in',width=10,height=10,res=300)
colnames(Corr) = rep('',15)
rownames(Corr) = rep('',15)
corrplot(abs(Corr), tl.col="black", 
         tl.srt=0.001, cl.offset = 1, tl.cex=1.2, col.lim = c(0,1),cl.cex = 1.5,
         col= colorRampPalette(c("midnightblue","white","darkred"))(100), 
         outline=TRUE, addgrid.col="white", type="lower", diag=T)
dev.off()
```
