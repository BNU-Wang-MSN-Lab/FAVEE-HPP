```r
library(ggdendro) #get the details
library(dendextend) #change the plot

library(tidyverse)
library(MASS)#MDS
#library(Rtsne)#tsne

library(ggrepel)
library(ape)#unrooted
```

```r
Lexical_dist = read.csv('rdm/Lexical_dist.csv',row.names=1)

options(repr.plot.width=20,repr.plot.height=10)
par(mar=c(20,1,1,1))
dend = Lexical_dist %>% 
    as.dist() %>%
    hclust(method = "complete") %>%
    as.dendrogram() %>%    
    set("labels_cex", 2) %>% 
    set("branches_lwd", 8) %>%
    set("branches_k_color",k=10)

plot(dend)
plot(dend,type='triangle')
```
