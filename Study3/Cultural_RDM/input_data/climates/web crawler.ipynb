```python
import requests
from pyquery import PyQuery
import re
import csv
```

```python
def get_html(url):
    """
    获取url对应的html数据（网页源代码字符串数据），返回html数据
    url: 网页的url
    """
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36 Edg/105.0.1343.50'}
    resp = requests.get(url,headers=headers)
    html = resp.text
    
    return html
```

```python
test = get_html('https://www.timeanddate.com/weather/australia/albury/climate')
```

```python
def extract_content(html):
    doc=PyQuery(html)
    climate_info = doc('#climateTable').text().split('一月')[0].split('\n')
    temp_list = []
    temp_list.append(climate_info[3].split('\xa0°C')[0])
    temp_list.append(climate_info[1].split('\xa0°C')[0])
    temp_list.append(climate_info[2].split('\xa0°C')[0])

    temp_dict = {i.split(': ')[0]:i.split(': ')[1] for i in temp_list}

    return temp_dict
```
