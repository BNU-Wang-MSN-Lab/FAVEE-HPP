```python
import pandas as pd
import numpy as np
import os

import time
import json

import openai
import tiktoken
from dotenv import load_dotenv
load_dotenv()
openai.api_key = os.getenv("GPT1_API") ##Extracting keys in the default environment. GPT-4.0
```

```python
def askChatGPT(messages):
    MODEL = os.getenv("MODEL1") 
    response = openai.ChatCompletion.create(
        model=MODEL,
        messages=messages,
        temperature=0) # for replication
    return response['choices'][0]['message']['content']  # type: ignore


def get_arr(a):
    init = [{
        "role": "user",
        "content": a
    }]
    return init
```

```python
CHN_rels_df  = pd.read_csv('CHN_modern/labels_chinese.csv', encoding = 'gb18030')
CHN_modern_258r = list(CHN_rels_df.loc[:,'query'])
CHN_modern_258r[0:5]
```

```python
askChatGPT(get_arr('你好！'))
```

```python
# Iterating through the list of relationships in groups of five.
for i in range(0, 5, 5):
    # Fetching the list of relationships for the current group.
    relationships = CHN_modern_258r[i:i+5]

    # Create prompt
    prompt = """请你在当代中国的社会文化背景下，描述一些人际关系属性，呈现它的内涵和特点。现在我将提供关系列表，请你依次返回对它们关系性质的描述：\n"""
    prompt += '\n'.join(relationships)
    prompt += "\n----\n请返回json格式结果，以关系名为key，关系描述为value，如{‘父母和孩子’:{‘描述’:内容}}。"

    print(prompt)
    
    # Obtaining the results.
    result = askChatGPT(get_arr(prompt))
    with open(os.path.join('CHN_modern/archieve/example.txt'),'w',encoding='utf-8') as file_group:
         file_group.write(result)
```

```python
files = os.listdir('CHN_modern/archieve/')
df = pd.DataFrame({'prompt':[],'relation':[],'description':[]})
for i in range(0, len(CHN_modern_258r), 5):
    relationships = CHN_modern_258r[i:i+5]
    archive_path = 'CHN_modern/archieve/'+str(i)+'.txt'
    with open(archive_path, 'r', encoding= 'utf-8') as f:
        archieve_results = json.load(f)
        temp_df = pd.DataFrame(archieve_results).transpose()
        temp_df.columns = ['description']
        temp_df.insert(0,'relation',temp_df.index)
        temp_df.insert(0,'prompt',relationships)
        #df = df.append(temp_df, ignore_index=True)
        df = pd.concat([df, temp_df],ignore_index=True)
df.to_csv('CHN_modern/chn_modern_results.csv',encoding='gb18030')
```
